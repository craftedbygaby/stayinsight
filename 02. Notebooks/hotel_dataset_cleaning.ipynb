Cleaning of the tourism_hospitality_data and transformation into two csv: 
    - One filtered to only Paris
    - One for all the cities but with all the countries
    

```python
import pandas as pd
import numpy as np
import re
from pathlib import Path

file_path = Path(r"C:\Users\johan\Desktop\hotel_pricing_project\data\raw\Tourism_Hospitality_Industry_Analysis.csv")

df = pd.read_csv(file_path)

print("Archivo cargado correctamente")
print("Shape:", df.shape)
display(df.head())
print("\nColumnas:")
print(df.columns.tolist())
```

```python

df_clean = df.copy()


df_clean.columns = [col.strip().lower() for col in df_clean.columns]
df_clean.columns = [re.sub(r"[^a-z0-9]+", "_", col).strip("_") for col in df_clean.columns]


text_cols = df_clean.select_dtypes(include="object").columns

for col in text_cols:
    df_clean[col] = df_clean[col].astype(str).str.strip()


print("Shape:", df_clean.shape)

print("\nNulos por columna:")
print(df_clean.isna().sum())

print("\nDuplicados:", df_clean.duplicated().sum())

print("\nColumnas limpias:")
print(df_clean.columns.tolist())

display(df_clean.head())
```

```python
# 1. Crear nombre del mes
month_map = {
    1: "January", 2: "February", 3: "March", 4: "April",
    5: "May", 6: "June", 7: "July", 8: "August",
    9: "September", 10: "October", 11: "November", 12: "December"
}

df_clean["month_name"] = df_clean["month"].map(month_map)

# 2. Convertir hotel_rating a numérico sin warning
rating_map = {
    "1-star": 1,
    "2-star": 2,
    "3-star": 3,
    "4-star": 4,
    "5-star": 5,
    "unrated": np.nan,
    "Unrated": np.nan
}

df_clean["hotel_rating_num"] = df_clean["hotel_rating"].map(rating_map)

# 3. Filtrar solo París
df_paris = df_clean[
    (df_clean["city"].str.lower() == "paris") &
    (df_clean["country"].str.lower() == "france")
].copy()

# 4. Comprobaciones
print("Shape df_clean:", df_clean.shape)
print("Shape df_paris:", df_paris.shape)

print("\nValores únicos de hotel_rating:")
print(sorted(df_clean["hotel_rating"].unique()))

print("\nPrimeras filas de df_paris:")
display(df_paris.head())
```

```python
from pathlib import Path

output_folder = Path("../data/processed")
output_folder.mkdir(parents=True, exist_ok=True)


df_clean.to_csv(output_folder / "tourism_hospitality_cleaned.csv", index=False)
df_paris.to_csv(output_folder / "paris_tourism_hospitality_cleaned.csv", index=False)

print("File saved in:")
print((output_folder / "tourism_hospitality_cleaned.csv").resolve())
print((output_folder / "paris_tourism_hospitality_cleaned.csv").resolve())
```
